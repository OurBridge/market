```python
# 전체 데이터 불러오기

import requests

def fetch_data():
    url = "https://api.odcloud.kr/api/15052837/v1/uddi:5bfc7a27-a1f8-4d86-a211-a7a02c0739d7"
    per_page = 100  # 페이지당 개수
    total_records = total_count  # 전체 데이터 개수
    total_pages = (total_records + per_page - 1) // per_page  # 총 페이지 수

    data = []

    for page in range(1, total_pages + 1):
        params = {
            "page": page,
            "perPage": per_page,
            "serviceKey": "CDKWiYrOyT0XapzcNW1DIE3exiXZz7XIpAIrrsOC/rkFLM1niU7Ukkn0Hg8Cfe/lepOAe/jJeHlTHSroOSOgYg=="
        }

        try:
            response = requests.get(url, params=params)
            json_data = response.json()
            market_data = json_data["data"]

            data.extend(market_data)
        except requests.exceptions.RequestException as e:
            print("Error fetching data:", e)
            return []

    return data

# fetch_data 함수 호출
result = fetch_data()

# '점포수'가 store개 이상인 데이터만 필터링하여 새로운 리스트에 저장
filtered_result = [item for item in result if item['점포수'] >= store]

```

```python
# 위도, 경도 추출하는 코드

import requests
import uuid

def get_geocode(address):
    url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode"
    headers = {
        "X-NCP-APIGW-API-KEY-ID": "mxphjh1u0c",
        "X-NCP-APIGW-API-KEY": "WlXAlmMvaB1CJzc7qSkuzO67RAGD5TtXrMycpTrb"
    }
    params = {
        "query": address
    }

    try:
        response = requests.get(url, headers=headers, params=params)
        json_data = response.json()
        geocode_data = json_data.get("addresses", [])

        if geocode_data:
            first_result = geocode_data[0]
            geocode = {
                "latitude": first_result["y"],
                "longitude": first_result["x"]
            }
            return geocode
    except requests.exceptions.RequestException as e:
        print("Error fetching geocode data:", e)
    
    return None

# fetch_data 함수 호출
result = fetch_data()

addresses = []
addresses_ = []
geocodes = []
markets = []
open_dates = []
market_types = []
scales = []
market_items = []
homepages = []
cities = []
uids = []

# '점포수'가 200개 이상인 경우만 필터링하여 데이터 추출 및 지리정보 가져오기
for item in result:
    if item['점포수'] >= store:
        uid = str(uuid.uuid4())
        address = item.get("주소(지번)")
        address_ = item.get("주소(도로명)")
        market = item.get("시장(상점가명)")
        open_date = item.get("시장개설주기")
        market_type = item.get("시장유형")
        market_item = item.get("취급품목")
        scale = item.get("점포수")
        homepage = item.get('홈페이지주소')
        if address:
            si = address.split()[0]
            if si == '서울특별시':
                si = '서울'
            elif si == '부산광역시':
                si = '부산'
            elif si == '대구광역시':
                si = '대구'
            elif si == '인천광역시':
                si = '인천'
            elif si == '광주광역시':
                si = '광주'
            elif si == '대전광역시':
                si = '대전'
            elif si == '울산광역시':
                si = '울산'
            elif si == '세종특별자치시':
                si = '세종'
            elif si == '제주특별자치도':
                si = '제주'
            else:
                si = si
            uids.append(uid)
            geocode = get_geocode(address)
            addresses.append(address)
            addresses_.append(address_)
            markets.append(market)
            geocodes.append(geocode)
            open_dates.append(open_date)
            market_types.append(market_type)
            scales.append(scale)
            market_items.append(market_item)
            homepages.append(homepage)
            cities.append(si)
            

# addresses 배열과 geocodes 배열에 저장된 데이터 출력
for i in range(len(addresses)):
    print("uid:", uids[i])
    print("지번 주소:", addresses[i])
    print("도로명 주소:", addresses_[i])
    print("시장:", markets[i])
    print("지리정보:", geocodes[i])
    print("시장개설주기:", open_dates[i])
    print("시장유형:", market_types[i])
    print("취급품목:", market_items[i])
    print("점포수:", scales[i])
    print("홈페이지주소:", homepages[i])
    print("시도군:", cities[i])
    print("-------------------")
```

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({ "uid": uids,
                    "시장정보": markets, 
                   "지번 주소": addresses, 
                   "도로명 주소": addresses_, 
                   "지리정보": geocodes, 
                   "시장개설주기": open_dates, 
                   "시장유형": market_types, 
                   "취급품목": market_items, 
                   "점포수": scales, 
                   "홈페이지주소": homepages,
                   "시도군": cities})

# 지리정보가 빈 값인 행 제거
df = df.dropna(subset=["지리정보"])

# CSV 파일로 저장 (한글 인코딩)
df.to_csv("result.csv", index=False, encoding="utf-8-sig")
```
