```python
# html 태그 삭제 전처리 + [title, link, contents]
import requests
from urllib.parse import quote
from bs4 import BeautifulSoup
import pandas as pd

# 전처리 함수: HTML 태그 제거
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    cleaned_text = soup.get_text()
    return cleaned_text

def call(keyword, start):
    encText = quote(keyword)
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&sort=sim&display=100&start=" + str(start)
    result = requests.get(url=url, headers={"X-Naver-Client-Id": "8HL0Kde0BuIxCob_IWH7", "X-Naver-Client-Secret": "KXphtnevh0"})
    return result.json()

def results(keyword):
    blog_results = []
    for num in range(0, 10):
        blog_results += call(keyword, num * 100 + 1)['items']
    return blog_results

keyword = "양동시장"
blog_results = results(keyword)

# Create lists to store the extracted data
titles = []
links = []
contents = []

# Extract 'title', 'link', and 'description' from each blog result
for result in blog_results:
    title = result['title']
    link = result['link']
    content = result['description']
    
    # 전처리: HTML 태그 제거
    title = remove_html_tags(title)
    content = remove_html_tags(content)
    
    titles.append(title)
    links.append(link)
    contents.append(content)

# Create a pandas DataFrame with the extracted data
df = pd.DataFrame({'제목': titles, '링크': links, '내용': contents})

# Save the DataFrame as a CSV file with UTF-8 encoding
df.to_csv('blog_search_results.csv', encoding='utf-8-sig', index=False)

# Print the DataFrame
print(df)
```

```python
from wordcloud import WordCloud
from konlpy.tag import Twitter
from collections import Counter
import matplotlib.pyplot as plt

# open으로 txt파일을 열고 read()를 이용하여 읽는다.
text = open('results.txt', encoding='utf-8').read()

twitter = Twitter()

# twitter함수를 통해 읽어들인 내용의 형태소를 분석한다.
sentences_tag = []
sentences_tag = twitter.pos(text)

noun_adj_list = []

# Define your remove_keywords list
remove_keywords = ['말바', '우시', '안녕하세요', '서구', keyword, '시장', '오늘', '광주', '광주시', '곳', '양동', '경안', '경기도', '경기', '화재', '이번', '봉', '위해', '입니다', '광역시', '있는', '위치', '치아', '때', '방문', '주차', '집', '수', '추천', '바로', '임플란트', '광고', '것', '번', '내']

# tag가 명사이거나 형용사이면서 remove_keywords에 속하지 않는 단어들만 noun_adj_list에 넣어준다.
for word, tag in sentences_tag:
    if tag in ['Noun', 'Adjective'] and word not in remove_keywords:
        noun_adj_list.append(word)

# 가장 많이 나온 단어부터 N개를 저장한다.
counts = Counter(noun_adj_list)
tags = counts.most_common(100)
print(tags)
```

```python
# 전처리 함수: HTML 태그 제거 및 단어 형태로 저장

import requests
from urllib.parse import quote
from bs4 import BeautifulSoup
import pandas as pd
from collections import Counter
import re

def preprocess_text(text):
    # HTML 태그 제거
    soup = BeautifulSoup(text, 'html.parser')
    cleaned_text = soup.get_text()
    
    # 단어 형태로 저장
    words = re.findall(r'\b\w+\b', cleaned_text.lower())
    return words

def call(keyword, start):
    encText = quote(keyword)
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&sort=sim&display=100&start=" + str(start)
    result = requests.get(url=url, headers={"X-Naver-Client-Id": "8HL0Kde0BuIxCob_IWH7", "X-Naver-Client-Secret": "KXphtnevh0"})
    return result.json()

def results(keyword):
    blog_results = []
    for num in range(0, 10):
        blog_results += call(keyword, num * 100 + 1)['items']
    return blog_results

keyword = "말바우시장"
blog_results = results(keyword)

# Create a list to store the extracted data
data = []

# Extract 'title' and 'description' from each blog result
for result in blog_results:
    title = result['title']
    description = result['description']
    
    # 전처리: HTML 태그 제거 및 단어 형태로 저장
    title_words = preprocess_text(title)
    description_words = preprocess_text(description)
    
    data.append({'title': title_words, 'description': description_words})

# Convert the list to a pandas DataFrame
df = pd.DataFrame(data)

# Function to count word frequencies
def count_word_frequencies(data, column_name):
    word_counts = Counter()
    for words in data[column_name]:
        word_counts.update(words)
    return word_counts

# Count word frequencies in 'title' column
title_word_frequencies = count_word_frequencies(df, 'title')

# Count word frequencies in 'description' column
description_word_frequencies = count_word_frequencies(df, 'description')

# Print the word frequencies
print("Title Word Frequencies:")
print(title_word_frequencies.most_common())
print("\nDescription Word Frequencies:")
print(description_word_frequencies.most_common())
```
