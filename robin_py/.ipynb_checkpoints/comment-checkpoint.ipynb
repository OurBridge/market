```python
import json
import uuid
import random

data = []

tags = ["가족과 함께", "친구와 함께", "커플", "혼자 여행", "맛집", "관광지", "문화", "쇼핑", "자연", "역사"]
markets = ["대인시장", "양동시장", "서부시장", "말바우시장"]

for _ in range(100):
    uid = str(uuid.uuid4())
    market = random.choice(markets)
    id = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz') for _ in range(8))
    comment = random.choice(["가족이랑 다녀왔어요. 먹을 게 많아서 좋아요.", 
                             "친구들이랑 여행으로 방문한 곳, 다시 오고 싶어요.", 
                             "혼자 갔는데도 재미있었어요.", 
                             "맛집이 많아서 식욕이 폭발했어요!",
                             "장도 보고 구경도 하고 즐거운 곳이에요.",
                             "외국인 친구가 가보고 싶다해서 왔어요. 시장 이모님들이 친절해서 친구에게 좋은 경험이 되었던 것 같아요.",
                             "마트보다 가격이 저렴해서 장보러 자주 와요.",
                             "시장 만의 정과 분위기가 너무 좋아서 계속 방문하고 싶어요.",
                             "여행하다 우연히 알게 되서 왔어요. 너무 좋아요."])
    img_url = random.choice(["https://minio.nculture.org/amsweb-opt/multimedia_assets/31/24894/7063/c/24894-medium-size.jpg",
    "https://www.kyongbuk.co.kr/news/photo/201906/2006493_410512_0109.jpg",
    "https://korean.visitseoul.net/comm/getImage?srvcId=MEDIA&parentSn=24165&fileTy=MEDIA&fileNo=1",
    "https://www.kyongbuk.co.kr/news/photo/201904/1059223_342424_0642.jpg",
    "https://visitgangnam.net/wp-content/uploads/2022/08/2-%E1%84%8B%E1%85%A7%E1%86%BC%E1%84%83%E1%85%A9%E1%86%BC%E1%84%8C%E1%85%A5%E1%86%AB%E1%84%90%E1%85%A9%E1%86%BC%E1%84%89%E1%85%B5%E1%84%8C%E1%85%A1%E1%86%BC-1-uai-1920x1080.jpg"])
    tag = random.sample(tags, random.randint(1, 5))
    
    json_obj = {
        "uid": uid,
        "시장정보": market,
        "id": id,
        "comment": comment,
        "img_url": img_url,
        "tag": tag
    }
    
    data.append(json_obj)

json_data = json.dumps(data, ensure_ascii=False, indent=2)
print(json_data)
```
